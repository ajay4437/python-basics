# Inheritance

It is nothing but a parent-child releation.
Inheritance: Aquering properties of parent class.
Parent is a base class.
Child is a derived class.
If we create an object for the child class, it can we used to access the attributes and function of the parent class

```python
# with out Inheritance:
class Baseclass:
    a=10
    b=100
    def display(self):
        print('base class')
class Derivedclass:
    c=20
    d=30
    def show(self):
        print("derived class")

a1 = Baseclass()
a2 = Derivedclass()
print(a1.a,a2.c)
a1.display()

print(a2.c, a2.d)
a2.show()
```

```python
# With Inheritance:
class Baseclass: # Parent class
    a=10
    b=100
    def display(self):
        print('base class')
class Derivedclass(Baseclass): # child class
    c=20
    d=30
    def show(self):
        print("derived class")


a2=Derivedclass() # Object of child class       

print(a2.a, a2.b) # Object of child class can be used to access the attributes and functions of parent class

a2.display()

print(a2.c, a2.d)
a2.show()
```

###### Note: we can not access the derived class object by the base class object.

Single Inheritance:
It has one base class and a derived class

Multilevel Inheritance: Base class(Grand-parent class)----->Derived class(Parent class)------>Derived class(child class)

```python
class grandfather:
    a=10
    def print1(self):
        print('grandfather')
    
class father(grandfather):
    b=20
    def print2(self):
        print('father')
            
class child(father):
    c=30
    def print3(self):
        print('child')
                    
ajay=child()

print(ajay.a)
print(ajay.b)
print(ajay.c)

ajay.print1()
ajay.print2()
ajay.print3()
```

### hierarchical inheritance:

It has one Bass class and two Derived class, which can access the attributes and methods of Base class and it self but not the other derived class

```python
class father_1:
    d=10
    def print4(self):
        print('father')
    
class child_1(father_1):
    e=20
    def print5(self):
        print('child_1')
        
class child_2(father_1):
    f=30
    def print6(self):
        print('child_2')
                    
a1=child_1()
a2=child_2()

print(a1.e)
print(a2.f)
print(a1.d)
print(a1.d)

a1.print4()
a1.print5()

a2.print4()
a2.print6()
```

### Multiple Inheritance:

It has two base class and a single Derived class. In simple words Mother and Father are base class,  where as child is a Derived class.

```python
class father:
    def print1(self):
        print('father')
class mother:
    def print2(self):
        print('mother')
class child(father, mother):
    def print3(sef):
        print('chid')

a=child()
a.print3()
a.print2()
a.print1()
        
```

# Polimorphism:

Implementing same thing in different ways. There are two types of polimorphism. 1.Compile Time  2.Run Time

### 1.Compile Time Polimorphism (Method Over loading)

 This can be done using "Method Overloading".

```python
class demo:
    def add(self,a,b=500,c=700): # This is known as default parameters
        print(a+b+c)
obj=demo()
obj.add(100) # This has only one parameter, so b=500, c=700 will be taken as default parameter
obj.add(100,200) # this has two parameters, so c=700 will be taken as default parameter
obj.add(100,200,300)
```

#### Note:
Default parameters ashould  be at the last.

### 2. Run Time Polimorphism (Method Over Riding)

In this case Base Class and Derived Class should have the same method name(function) and having same parameters in it. We will initiate an object for the child class. When we access the Parent Class method using the child class object, if it is not satisified then it is over-ride by the child class method

```python
class parent:
    def transport(self):
        print('cycle')
class child(parent):
    def transport(self):
        print('Bike')
obj=child()
obj.transport() # By default parent method is over-ride by the child method, so the out put is 'bike'
```

# Abstraction:
   Hiding the implentation and showing the essential part only.
   
   To do the Abstraction, we have to import a module known as "abc" module which stands for Abstract Base class. It has "abc"      class.
   
   Any class is said to be abstract class, when it has atleast one abstraction method.
   
   We cannot create an object to an abstract class. To create an object we have to convert abstract class to concrete class (having abstract methods) 

#### Abstract Method:
     it has only function call with empty defination.
     To impliment this we need a decorator
     The decorator is "@abstract method"

#### Note:
     To create  an object we have to over-ride the abstract class with the concrete class.
        We consider the abstract class as the Base class and concrete class as the Derived class and we override the methods of the abstract class with the concrete class, so that we can create an object to the concrete class and use the methods of the abstract class
     

```python
from abc import ABC,abstractmethod
class abstractdemo(ABC):             # Abstract class
    @abstractmethod
    def housing_intrest(self):  # Abstract method
        None
    @abstractmethod
    def vehicle_intrest(self): # Abstract method
        None
class sbi(abstractdemo):        # Concrete class
    def housing_intrest(self): 
        print('housing intrest = 8.5%')
    def vehicle_intrest(self): 
        print('vehicle intrest = 7.5%')
obj=sbi()
obj.housing_intrest()
obj.vehicle_intrest()
    
```

#### Note:
    No of Abstract methods in abstract class SHOULD BE EQUAL to the methods in concrete class

# Data Hiding:
    giving security to the data, by access specifiers
    

```python
#data hiding:
class demo:
    a=10
    b=20
    def display(self):
        print('Data hiding')
obj=demo()
obj.display()
print(obj.a, obj.b)
```

```python
# outside the class we cannot access the private variable:
class demo:
    __a=10 # This is the representation of private variable
    b=20
    def display(self):
        print('Data hiding')
obj=demo()
obj.display()
print(obj.__a, obj.b)
```

#### Note: 
    outside the class we cannot access the private variable. but we can access the private variable within the class

```python
# we can access the private variable within the class
class demo:
    __a=10
    b=20
    def display(self):
        print(self.__a)
        print('Data hiding')
obj=demo()
obj.display()
print(obj.b)
```

##### Note:
    It is same with the methods.

```python
class demo:
    __a=10
    b=20
    def __display(self): #this is the private method
        print(self.__a)
        print('Data hiding')
    def show(self): # to call the private method, we created a public method
        self.__display() # calling the private method
    
obj=demo()
obj.show()
print(obj.b)
```
